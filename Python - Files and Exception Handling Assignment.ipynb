```python
'''Q.15] How does Python's garbage collection system work?
Ans. Python's garbage collection (GC) system is responsible for managing memory by reclaiming unreachable objects 
(objects no longer referenced by the program) and ensuring efficient memory utilization. It is an integral part of Python’s memory management system.

Key Features of Python's Garbage Collection System

1. Automatic Memory Management:
Python automatically allocates and deallocates memory to objects as needed during program execution.

2. Reference Counting Mechanism:
Python primarily uses reference counting to track objects. Each object in Python has a reference count, which is the number of references pointing to it.
When an object’s reference count drops to zero (i.e., it is no longer reachable), the memory occupied by the object is immediately deallocated.

3. Handling Cyclic References:
Reference counting alone cannot handle cyclic references, where two or more objects reference each other but are otherwise unreachable.
To address this, Python employs a cyclic garbage collector, which identifies and collects cyclic references.'''
print("For Example :")
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

a = Node(1)
b = Node(2)
a.next = b
b.next = a  # Creates a cycle: a -> b -> a

del a
del b  # Objects are unreachable but not collected by reference counting

'''In such cases, the garbage collector detects the cycle and reclaims the memory.

4. Generational Garbage Collection:
Python’s garbage collector uses a generational approach to optimize performance. Objects are divided into three generations:
Generation 0: Newly created objects (young generation).
Generation 1: Objects that have survived a collection from Generation 0.
Generation 2: Objects that have survived multiple collections (long-lived objects).
The idea is that most objects die young, so Generation 0 is collected more frequently than Generation 1 and 2.

5. Garbage Collection Triggers:
The garbage collector is triggered automatically when certain thresholds are exceeded (e.g., the number of allocations exceeds a limit).
You can also manually invoke garbage collection using the gc module.'''
print("For Example :")
import gc
gc.collect()  # Manually triggers garbage collection

'''6. Customizing Garbage Collection Behavior:
The gc module allows developers to control garbage collection, such as enabling/disabling it or tuning thresholds.'''
print("For Example :")
import gc
gc.disable()  # Disables garbage collection
gc.enable()   # Enables garbage collection
```

```python
'''Q.23] Write a Python program that writes to a log file when an error occurs during file handling.'''
import logging

# Configure logging
logging.basicConfig(
    filename="file_handling_errors.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def read_file(file_path):
    try:
        with open(file_path, "r") as file:
            content = file.read()
            print("File content:")
            print(content)
    except FileNotFoundError:
        logging.error(f"The file '{file_path}' was not found.")
        print(f"Error: The file '{file_path}' does not exist.")
    except PermissionError:
        logging.error(f"Permission denied while trying to read the file '{file_path}'.")
        print(f"Error: Permission denied for file '{file_path}'.")
    except Exception as e:
        logging.error(f"An unexpected error occurred: {e}")
        print(f"Error: An unexpected error occurred: {e}")

    with open("file_handling_errors.log", "r") as file2:
        print(file2.read())

# Example usage
file_path = "example4.txt"  # Replace with your file path
read_file(file_path)
```
